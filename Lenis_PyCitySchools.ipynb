```python
# Solution to the Academy of Py problem

# Importin dependencies 
import pandas as pd
import numpy as np
```

```python
# Setting the paths
path_school = "PyCitySchools/Resources/schools_complete.csv"  # Please update this object accordingly
path_students = "PyCitySchools/Resources/students_complete.csv"  # Please update this object accordingly
```

```python
PASSING_GRADE_MATH = 65 # Change this accordingly
PASSING_GRADE_READ = 65 # Change this accordingly
```

```python
# School level data
schools = pd.read_csv(path_school).sort_values(by = ['school_name'])
schools.head()
```

```python
#Checking that there are not repeated observations
number_different_school_ids = len(schools["School ID"].unique())
number_rows_school_df       = len(schools["School ID"])

if number_different_school_ids == number_rows_school_df:
    print("Each row is associated with a unique school")
```

```python
# Students Data
students = pd.read_csv(path_students).sort_values(by = ['school_name'])
students.head()
```

### Question 1: District Summary

* Create a high level snapshot (in table form) of the district's key metrics, including:
  * Total Schools
  * Total Students
  * Total Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Checking if btoh dataframes have the same schools
schools_list_from_schools  = sorted(list(schools["school_name"]))
schools_list_from_students = sorted(list(students["school_name"].unique()))

if schools_list_from_schools == schools_list_from_students:
    print("Both dataframes include the same schools names, you can do a inner join")
```

```python
students_and_schools = pd.merge(students, schools, how = 'inner', on = "school_name").sort_values(by = ['school_name'])
students_and_schools.head()
```

```python
# I am assuming that the passing grade is 65
students_and_schools["pass_reading"] = [100 if x >= PASSING_GRADE_READ else 0 for x in students_and_schools["reading_score"]]
students_and_schools["pass_math"] = [100 if x >= PASSING_GRADE_MATH else 0 for x in students_and_schools["math_score"]]
students_and_schools.head()
```

```python
# I am assuming that, for the first question, we care only about the "District" type of schools, since the 
# question says "District Summary" however the solution seems to include information from charter and 
# district schools
students_and_schools_districts = students_and_schools[students_and_schools['type'] == 'District'].sort_values(by = ['school_name'])
students_and_schools_districts.head()
```

```python
district_summary_dict = {
    "Total Schools"         : len(students_and_schools_districts["School ID"].unique()), 
    "Total Students"        : format(students_and_schools_districts["size"].unique().sum(), ',d'),
    "Total Budget"          : '${:,.2f}'.format(students_and_schools_districts["budget"].unique().sum()),
    "Average Math Score"    : round(students_and_schools_districts["math_score"].mean(), 2),
    "Average Reading Score" : round(students_and_schools_districts["reading_score"].mean(), 2),
    "% Passing Math"        : round(students_and_schools_districts["pass_math"].mean(), 2),
    "% Passing Reading"     : round(students_and_schools_districts["pass_reading"].mean(), 2),
    "Overall Passing Rate"  : round(students_and_schools_districts["pass_math"].mean() * 0.5 + \
                                    students_and_schools_districts["pass_reading"].mean() * 0.5, 2)
}

district_summary = pd.DataFrame(district_summary_dict, index=[0])
district_summary
```

### Question 2: School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
q2_intermidiate = students_and_schools.groupby(students_and_schools['school_name']).mean().reset_index()
q2_intermidiate["overall_pass"] = q2_intermidiate["pass_math"] * 0.5 + q2_intermidiate["pass_reading"] * 0.5
q2_intermidiate = q2_intermidiate[["school_name" , "reading_score", "math_score", 
                                   "pass_reading", "pass_math"    , "overall_pass"]]
q2_intermidiate = q2_intermidiate.sort_values(by = ['school_name'])
q2_intermidiate
```

```python
summary_q2 = schools.sort_values(by = ['school_name'])
summary_q2["budget_per_student"] = summary_q2["budget"] / summary_q2["size"]
summary_q2 = pd.merge(summary_q2, q2_intermidiate, how = "left", on = "school_name")

summary_q2
```

```python
# Selecting the variables we care about
final_q2 = summary_q2[["school_name"       , "type"      , "size"         , "budget"   , 
                       "budget_per_student", "math_score", "reading_score", "pass_math",
                       "pass_reading"      , "overall_pass"]]

final_q2.columns = ["School Name"           , "School Type"       , "Total Students"    , 
                    "Total School Budget"   , "Per Student Budget", "Average Math Score",
                    "Average Reading Score" , "% Passing Math"    , "% Passing Reading" ,
                    "% Overall Passing Rate"]

final_q2 = final_q2.set_index("School Name")

# Making things look pretty
final_q2["Total School Budget"]    = final_q2["Total School Budget"].apply(lambda x: "${:,.2f}".format(x))
final_q2["Per Student Budget"]     = final_q2["Per Student Budget"].apply(lambda x: "${:,.2f}".format(x))
final_q2["Average Math Score"]     = round(final_q2["Average Math Score"], 2)
final_q2["Average Reading Score"]  = round(final_q2["Average Reading Score"], 2)
final_q2["% Passing Math"]         = round(final_q2["% Passing Math"], 2)
final_q2["% Passing Reading"]      = round(final_q2["% Passing Reading"], 2)
final_q2["% Overall Passing Rate"] = round(final_q2["% Overall Passing Rate"], 2)

final_q2
```

### Question 7: Scores by School Size
* Perform the same operations as above, based on school size.


```python
# I am going to use the bins provided
size_bins = [0, 1000, 2000, 5000]
size_bins_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

df_binned_by_school_size = students_and_schools
df_binned_by_school_size["school_size"] = pd.cut(df_binned_by_school_size["size"], 
                                                 bins = size_bins, 
                                                 labels = size_bins_names)

df_binned_by_school_size.head()
```

```python
df_q7 = df_binned_by_school_size.groupby(["school_size"]).mean().reset_index().round(2)
df_q7["overall_pass"] = df_q7["pass_math"] * 0.5 + df_q7["pass_reading"] * 0.5
df_q7 = df_q7[["school_size", "math_score", "reading_score", "pass_math", "pass_reading", "overall_pass"]]

df_q7.columns = ["School Size"   , "Average Math Score", "Average Reading Score", 
                 "% Passing Math", "% Passing Reading" , "% Overall Passing Rate"]

df_q7_final = df_q7.set_index("School Size")

df_q7_final
```

### Question 8: Scores by School Type

```python
df_q8 = df_binned_by_school_size.groupby(["type"]).mean().reset_index().round(2)
df_q8["overall_pass"] = df_q8["pass_math"] * 0.5 + df_q8["pass_reading"] * 0.5
df_q8 = df_q8[["type", "math_score", "reading_score", "pass_math", "pass_reading", "overall_pass"]]

df_q8.columns = ["School Type"   , "Average Math Score", "Average Reading Score", 
                 "% Passing Math", "% Passing Reading" , "% Overall Passing Rate"]

df_q8_final = df_q8.set_index("School Type")
df_q8_final
```
